```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab



file = open('lwd_travel_distance.txt')
traveld = file.readlines()
data = []
for a in range(len(traveld)):
    # remove linebreak which is the last character of the string
    data.append(float(traveld[a][:-1]))

print(data)
print(len(data))
```

```python
nozero = []
for x in data:
    if x >= 0.2:
        nozero.append(x)
print(nozero)
yll = 30
num_bins = 10
n, bins, patches = plt.hist(nozero, num_bins, facecolor='gray', edgecolor = 'black', alpha=0.5)
plt.xlabel('travel distance [rkm]')
plt.ylabel('# of pieces of LWD')
plt.title(str(num_bins) + ' bins')
plt.ylim(0, yll)
plt.show()

num_bins = 15
n, bins, patches = plt.hist(nozero, num_bins, facecolor='gray', edgecolor = 'black', alpha=0.5)
plt.xlabel('travel distance [rkm]')
plt.ylabel('# of pieces of LWD')
plt.title(str(num_bins) + ' bins')
plt.ylim(0, yll)
plt.show()

num_bins = 25
n, bins, patches = plt.hist(nozero, num_bins, facecolor='gray', edgecolor = 'black', alpha=0.5)
plt.xlabel('travel distance [rkm]')
plt.ylabel('# of pieces of LWD')
plt.title(str(num_bins) + ' bins')
plt.ylim(0, yll)
plt.show()

num_bins = 50
n, bins, patches = plt.hist(nozero, num_bins, facecolor='gray', edgecolor = 'black', alpha=0.5)
plt.xlabel('travel distance [rkm]')
plt.ylabel('# of pieces of LWD')
plt.title(str(num_bins) + ' bins')
plt.ylim(0, yll)
plt.show()

num_bins = 75
n, bins, patches = plt.hist(nozero, num_bins, facecolor='gray', edgecolor = 'black', alpha=0.5)
plt.xlabel('travel distance [rkm]')
plt.ylabel('# of pieces of LWD')
plt.title(str(num_bins) + ' bins')
plt.ylim(0, yll)
plt.show()
```
